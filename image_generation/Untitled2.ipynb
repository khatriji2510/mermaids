```python
TOTAL_MERMAIDS = 9999

traits = []

def createCombo():
    
    trait = {}

    trait["Background"] = random.choices(backgrounds, backgroundweights)[0]
    trait["Skin"] = random.choices(Skin, Skinweights)[0]
    trait["Lips"] = random.choices(Lips, Lipsweights)[0]
    trait["Eyes"] = random.choices(eyes, eyesweights)[0]
    trait["Head accessories"] = random.choices(HeadAce, HeadAceweights)[0]
    trait["Fin"] = random.choices(Fin, Finweights)[0]
    trait["Hair"] = random.choices(Hair, Hairweights)[0]
    trait["Top"] = random.choices(Top, Topweights)[0]
    
    if trait in traits:
        return createCombo()
    else:
        return trait
    
for i in range(TOTAL_MERMAIDS):
    
    newtraitcombo = createCombo()
    
    traits.append(newtraitcombo)
```

```python
def allUnique(x):
    seen = list()
    return not any(i in seen or seen.append(i) for i in x)

print(allUnique(traits))
```
